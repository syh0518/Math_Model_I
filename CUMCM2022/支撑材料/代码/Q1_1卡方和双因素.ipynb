```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
Q1_1 = pd.read_excel("Q1_1卡方.xlsx")
Q1_1
```

```python
count = pd.crosstab(index=Q1_1['表面风化'],columns=Q1_1['类型'])
count
```

```python
x = [[12,12],[24,6]]
df = pd.DataFrame(x,index=['无风化','风化'],columns=['铅钡','高钾'])
chi2, p_value,df,f_exp=chi2_contingency(df)
print(f"卡方统计量={chi2:.3f}, 自由度={df} ,p值={p_value:.4g}。\n期望频数为: \n{f_exp}")
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
df1 = pd.read_csv("Q1_2方差.csv")
```

```python
model = ols('value~C(ad) + C(price)', df1).fit()
anova_lm(model)
```
